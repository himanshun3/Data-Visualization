```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('character-deaths.csv')
plt.subplot(1,2,1)
sns.boxplot(x='Death Year',y='Death Chapter',data=df)
plt.ylabel('Death Chapter')
plt.title('BoxPlot')
plt.subplot(1,2,2)
sns.violinplot(x='Death Year',y='Book Intro Chapter',data=df)
plt.ylabel('Death Chapter')
plt.title('Violin Plot')
plt.tight_layout()
plt.show()
```
